```python
from datasets import load_dataset
```

```python
import duckdb
```

```python
ds = load_dataset("marcelbinz/Psych-101", split="train")
```

```python
df = ds.to_pandas()
```

```python
con = duckdb.connect()

result_df = con.execute("""
with ranked_table as (
  select *,
  row_number() over (partition by experiment order by participant) as rn
  from df
)

select text, experiment, participant
from ranked_table
where rn =1
order by experiment asc;

""").fetchdf()

```

```python
result_df.to_csv("experiment_first_rows.csv", index= False)
```
